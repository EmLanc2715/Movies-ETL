```python
def clean_movie(movie):
    
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) >0:
        movie['alt_titles'] = alt_titles
        
        
    #Merge column names
    
    def change_column_name(old_name, new_name):
        
        if old_name in movie:
            
            movie[new_name] = movie.pop(old_name)
            
    change_column_name('Adapation by', 'Writer(s)')
    change_column_name('Country of origin', 'Country')
    change_column_name('Directed by', 'Director')
    change_column_name('Distributed by', 'Distributor')
    change_column_name('Edited by', 'Editor(s)')
    change_column_name('Length', 'Running time')
    change_column_name('Original release', 'Release date')
    change_column_name('Music by', 'Composer(s)')
    change_column_name('Produced by', 'Producer(s)')
    change_column_name('Producer', 'Producer(s)')
    change_column_name('Productioncompanies ', 'Production company(s)')
    change_column_name('Productioncompany ', 'Production company(s)')
    change_column_name('Released', 'Release Date')
    change_column_name('Release Date', 'Release date')
    change_column_name('Screen story by', 'Writer(s)')
    change_column_name('Screenplay by', 'Writer(s)')
    change_column_name('Story by', 'Writer(s)')
    change_column_name('Theme music composer', 'Composer(s)')
    change_column_name('Written by', 'Writer(s)')
    
    return movie
```

```python
import re
```

```python
# * says that the previous character can repeatany number of times

# ca*t would match "cat", "caaat", and "fiction"
```

```python
#If we want to specify the characters have to show up at least once, we use +

#"ca+t" would match "cat" and "caaat" but not "fiction"
```

```python
#If we wantto search for a character that shows up an exact number of typers use {}

# "ca{3,5}t" would not match cat or ficiton but would match "caaat"
```

```python
# Create a regex where our pattern match string will include six elements in the following order:
#a dollar sign, an arbitrary but nonsero number of digits,an optional decimal point, 
#an arbitrary but possibly zero number of more digits, a space maybe more than one, the word million or billion

form_one = r'\$\d+\.?\d*\s[bm]illion'

#Count how many box office values match the first form

box_office.str.contains(form_one, flags=re.IGNORECASE, na=False).sum()
```

```python
#Second form where we need our pattern to match a dollar sign, a group of one to three digits, at least one group
#starting with a comman and followed by exactl three digits

form_two = r'\$\d{1,3}(?:,\d{3})+'

box_office.str.contains(form_two, flags=re.IGNORECASE, na=False).sum()
```

```python
#Need to see which values arent described by form one and form two. 
#Create two Boolean series and then select which box office values dont match

matches_form_one = box_office.str.contains(form_one, flags=re.IGNORECASE, na=False)

matches_form_two = box_office.str.contains(form_two, flags=re.IGNORECASE, na=False)
```

```python
#See which values don't match either form

#This will make an error!

box_office[(not matches_form_one) and (not matches_form_two)]
```

```python
#Corrected code

box_office[~matches_form_one & ~matches_form_two]
```

```python
#Fix pattern matches

#1.Some values have spaces between the $ and number

form_one = 'r\$\s*\d+\.?\d*\s*[mb]illion'

#form_two = r'\$\s*\d{1,3}(?:,\d{3}+)

#2. SOme vlaues use a period as a thousands seperator and not a comma

form_two = r'\$\s*\d{1,3}(?:[,\.]\d{3})+'
```

```python
#3. Some values are given as a range
#To correct we are going to find any string that starts with a $ and ends with -. 
#Then replace the - with a $ using the replace() method

box_office = box_office.str.replace(r'\$.*[---](?!a-z)', '$', regex=True)
```

```python
#4. Million is sometimes misspelled as 'millon'
#To fix make the second i optional in the string match

form_one = r'\$s*\d+\.?\d*\s*[mb]illi?on'
```

```python
# Use str.extract to extract only the parts of the strings that match

#Use an f string to create a regex that matches either form one or two

box_office.str.extract(f'({form_one}|{form_two})')
```

```python
#Now we need a funtion to turn the extracted values into a numeric value
#We will call it parse_dollars and it will take a string and return a floating-point number
#use re.sub(pattern, replacement_string, string) to remove $, spaces, commas and letters if necessary
#Finally, convert all strings to floats, multiply by the right amount and retun the value

def parse_dollars(s):
    # if s is not a string, return NaN
    if type(s) != str:
        return np.nan

    # if input is of the form $###.# million
    if re.match(r'\$\s*\d+\.?\d*\s*milli?on', s, flags=re.IGNORECASE):

        # remove dollar sign and " million"
        s = re.sub('\$|\s|[a-zA-Z]','', s)

        # convert to float and multiply by a million
        value = float(s) * 10**6

        # return value
        return value

    # if input is of the form $###.# billion
    elif re.match(r'\$\s*\d+\.?\d*\s*billi?on', s, flags=re.IGNORECASE):

        # remove dollar sign and " billion"
        s = re.sub('\$|\s|[a-zA-Z]','', s)

        # convert to float and multiply by a billion
        value = float(s) * 10**9

        # return value
        return value

    # if input is of the form $###,###,###
    elif re.match(r'\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)', s, flags=re.IGNORECASE):

        # remove dollar sign and commas
        s = re.sub('\$|,','', s)

        # convert to float
        value = float(s)

        # return value
        return value

    # otherwise, return NaN
    else:
        return np.nan
```

```python
# Need to extract the values from box_office using str.extract then apply parse_dollars to the first column 
#of the DF returned by str.extract

wiki_movies_df['box_office'] = box_office.str.extract(f'({form_one}|{form_two})', flags=re.IGNORECASE)[0].apply(parse_dollars)
```

```python
#Create a budget variable 

budget = wiki_movies_df['Budget'].dropna()

#Convert any lists to strings

budget = budget.map(lambda x: ' '.join(x) if type(x) == list else x)

#Remove any values between a $ and -

budget = budget.str.replace(r'\$.*[-—–](?![a-z])', '$', regex=True)
```

```python
#Use the same pattern matches that I created to parse box office datat and apply without modicfication to budget data

matches_form_one = budget.str.contains(form_one, flags=re.IGNORECASE, na=False)

matches_form_two = budget.str.contains(form_two, flags=re.IGNORECASE, na=False)

budget[~matches_form_one & ~matches_form_two]
```

```python
#Remove citation references

budget = budget.str.replace(r'\[\d+\]\s*' , '')

budget[~matches_form_one & ~matches_form_two]
```

```python
#Copy code we used to parse the box office values, chagning box_office to budget

wiki_movies_df['budget'] = budget.str.extract(f'({form_one}|{form_two})', flags=re.IGNORECASE)[0].apply(parse_dollars)

#Drop the original budget column

wiki_movies_df.drop('Budget', axis=1, inplace=True)
```

```python
#Parse the Release Dates
#1. Make a variable that holds the non null values of Release date in DF, converting strings to lists

release_date = wiki_movies_df['Release date'].dropna().apply(lambda x: " ".join(x) if type(x) == list else x)
```

```python
#Forms we need to parse out
#1. Full month name, one to two digit day, four digit year
#2. Four digit year, two digit month, two digit day, with any seperator
#3. Full month name, four digit year
#4. Four digit year

date_form_one = r'(?:January|February|March|April|May|June|July|August|September|October|November|December)\s[123]?\d,\s\d{4}'

date_form_two = r'\d{4}.[01]\d.[0123]\d'

date_form_three = r'(?:January|February|March|April|May|June|July|August|September|October|November|December)\s\d{4}'

date_form_four = r'\d{4}'
```

```python
#Extract dates

release_date.str.extract(f'({date_form_one}|{date_form_two}|{date_form_three}|{date_form_four})', flags=re.IGNORECASE)
```

```python
#Use the built in to_datetime() method in Pandas instead of creating a new function to parse the dates

wiki_movies_df['release_date'] = pd.to_datetime(release_date.str.extract(f'({date_form_one}|{date_form_two}|{date_form_three}|{date_form_four})')[0], infer_datetime_format=True)
```

```python
#Parse Running Time
#Make a variable to hold the non null values of Running time in DF, converting to strings

running_time = wiki_movies_df['Running time'].dropna().apply(lambda x: " ".join(x) if type(x) == list else x)

running_time
```

```python
#Look at running times

running_time.str.contains(r'^\d*\s*minutes$', flags=re.IGNORECASE, na=False).sum()
```

```python
#Look at the other entries excluded by the above code

running_time[running_time.str.contains(r'^\d*\s*minutes$', flags=re.IGNORECASE, na=False) !=True]
```

```python
# Make it more general by only marking the neginning of the string and accepting abbreviations of minutes by searching m

running_time.str.contains(r'^\d*\s*m', flags=re.IGNORECASE, na=False).sum()
```

```python
#The remaining 17 follow:

running_time[running_time.str.contains(r'^\d*\s*m', flags=re.IGNORECASE, na=False) !=True]
```

```python
#Match all of the hour+ min patterns with on regex that needs to 
#1. Start with one or more digits
#2. Have an optional space after the digit and before the letter h
#3. Capture all possible abbreviations
#4. Have an optional space after the hours marker
#5. Have an optiona; number of digits

#"\d+\s*ho?u?r?s?\s*\d*"
```

```python
#Extract values with new pattern

running_time_extract = running_time.str.extract(r'(\d+)\s*ho?u?r?s?\s*(\d*)|(\d+)\s*m')

running_time_extract
```

```python
#Need to convert strings to numeric values

running_time_extract = running_time_extract.apply(lambda col: pd.to_numeric(col, errors='coerce')).fillna(0)

running_time_extract
```

```python
#Apply a function to convert the hour and minute capture groups if pure minutes capture group is zero, save to wiki_movies_df

wiki_movies_df['running_time'] = running_time_extract.apply(lambda row: row[0]*60 + row[1] if row[2] == 0 else row[2], axis =1)
```

```python
#Drop the running time column from the dataset

wiki_movies_df.drop('Running time', axis=1, inplace=True)
```
